### task 0
Write the function, that has argument "square's side lenth" as "a", and returns 2d array form a*a. The array must be filled with "1" and "0" in chessboard order.
* start with "0"
* use np.zeros 
* use slices
* no use cycles

```python
def get_chess(a):
    arr = np.zeros((a,a), dtype=np.uint8)
    arr[1::2, ::2] = 1
    arr[::2, 1::2] = 1
    return arr

get_chess(a=4)
```
