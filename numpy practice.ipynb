### task 0
Write the function, that takes argument "square's side lenth" as "a", and returns 2d array form a*a. The array must be filled with "1" and "0" in chessboard order.
* start with "0"
* use np.zeros 
* use slices
* no use cycles

```python
def get_chess(a):
    arr = np.zeros((a,a), dtype=np.uint8)
    arr[1::2, ::2] = 1
    arr[::2, 1::2] = 1
    return arr

get_chess(a=4)
```

### task 1
Write the function, that takes argument "array". The function must ganerate random number between 0 to uint32 for seed, and shafful items of array. The function returns a tuple with random number, and shaffuled array.
* the original array must remain unchanged
* the seed number must remain unchanged

```python
def shuffle_seed(array):
    rm_seed = np.random.randint(0,4294967295,dtype=np.int64)
    np.random.seed(rm_seed)
    return (np.random.permutation(array), rm_seed)

array = [1, 2, 3, 4, 5]
shuffle_seed(array)
```

### task 2
Write the function, that takes unlimited number of vectors as input. The function returns min and max distance between vectors, as a tuple.
* it is guaranteed, that all vectors are same lenth

bad solution

```python
def min_max_dist(*vectors):
    all_vec = [*vectors]
    result = list()
    for i in all_vec:
        for j in all_vec:
            if i is not j:
                result.append(np.linalg.norm(i - j))        
    return (np.min(result),np.max(result))

vec1 = np.array([1, 2, 3])
vec2 = np.array([4, 5, 6])
vec3 = np.array([7, 8, 9])

min_max_dist(vec1, vec2, vec3)
```

good solution

```python
def min_max_dist(*vectors):
    result = list()
    for i in range(len(vectors)):
        for j in range(i+1, len(vectors)):
            result.append(np.linalg.norm(vectors[i] - vectors[j]))
    return (np.min(result), np.max(result))

min_max_dist(vec1, vec2, vec3)
```

Write the function, that takes number of array parts as input, and generates 3d array of random numbers, between 1 to 100. The function returns a 5x5 array of random numbers.

### Task 5
Write the function, that takes number of array parts as input, and generates 3d array of random numbers, between 1 to 100. The function returns a 5x5 array of random numbers.
* numbers must be int
* numbers shouldn't repeated

```python
def get_unique_loto(num):
    i = 0
    lst_100 = np.arange(1, 101)
    arr = np.zeros((num,5,5))
    while i < num:
        arr[i,::,::] = np.random.choice(lst_100, size=(5,5), replace=False)
        i += 1    
    return arr
get_loto(num=3)
```
