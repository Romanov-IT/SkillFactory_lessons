# pandas sorting dataframe

```python
import pandas as pd
```

```python
melb = pd.read_csv("data/melb_data_fe.csv", index_col=False)
melb_df = melb.copy()
melb_df.head(3)
```

```python
melb_df["Date"] = pd.to_datetime(melb_df["Date"])
melb_df.info()
```

```python
melb_df["Date"].dt.quarter.value_counts()
```

```python
no_category = ["Date", "Rooms", "Bedroom", "Bathroom", "Car"]
for name_col in melb_df.columns:
    if (name_col not in no_category) and (melb_df[name_col].nunique() < 150):
        melb_df[name_col] = melb_df[name_col].astype("category")
melb_df.info()
```

### .sort_values()
* by= name or name list, that needs to sort
* axis=0 sorting by rows
* axis=1 sorting by column
* ascending=True as default sorting from smallest to largest
* ignore_index=False as default no add a new indexes
* inplace=False as default no replace original table

```python
melb_df.sort_values(by="Price").head(5)
```

```python
melb_df.sort_values(by="Date", ascending=False).head(5)
```

### sorting values of some columns

```python
melb_df.sort_values(by=["Distance", "Price"]).loc[::10, ["Distance", "Price"]]
```

```python
melb_df.sort_values(by=["Price", "Distance"]).loc[::10, ["Price", "Distance"]]
```

# sorting and filtering dataframe

```python
mask1 = melb_df["AreaRatio"] < 0.8 
mask2 = melb_df["Type"] == "townhouse"
mask3 = melb_df["SellerG"] == "McGrath"
melb_df[mask1 & mask2 & mask3].sort_values(
    by=["Date", "AreaRatio"],
    ascending=[True, False],
    ignore_index=True
).loc[:, ["Date", "AreaRatio"]]
```

```python
mask1 = melb_df["Type"] == "townhouse"
mask2 = melb_df["Rooms"] > 2
melb_df[mask1 & mask2].sort_values(
    by=["Rooms", "MeanRoomsSquare"],
    ascending=[True, False],
    ignore_index=True
).loc[18]
```
