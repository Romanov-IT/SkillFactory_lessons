# pandas grouping dataframe

```python
import pandas as pd
```

```python
melb_df = pd.read_csv("data/melb_data_fe.csv", index_col=False)
melb_df.head(5)
```

### .groupby()
* by= name or name list, that needs to groupe
* axis=0 as default groupe by rows
* axis=1 groupe by columns
* as_index=True as default add new index

### grouping data by one criterion with one aggregation

```python
melb_df.groupby(by="Type").mean(numeric_only=True)
```

```python
melb_df.groupby(by="Type", as_index=False).mean(numeric_only=True)
```

```python
melb_df.groupby("Type")["Price"].mean()
```

```python
melb_df.groupby("Regionname")["Distance"].min().sort_values(ascending=False)
```

### grouping data by one criterion with multiple aggregations

```python
melb_df.groupby("MonthSale")["Price"].agg(
    ["count", "mean", "max"]).sort_values(by="count", ascending=False)
```

```python
melb_df.groupby("MonthSale")["Price"].agg("describe").sort_values("count", ascending=False)
```

```python
melb_df.groupby("Regionname")["SellerG"].agg(["nunique", set]).sort_values("nunique", ascending=False)
```

```python
melb_df.groupby("Rooms")["Price"].mean().sort_values(ascending=False)
```

```python
melb_df.groupby("Regionname")["Lattitude"].std().sort_values(ascending=False)
```

```python
melb_df["Date"] = pd.to_datetime(melb_df["Date"])
```

```python
mask1 = (melb_df["Date"].dt.month > 4)
mask2 = (melb_df["Date"].dt.month < 10)
```

```python
melb_df[mask1 & mask2].groupby(by="SellerG")["Price"].sum().sort_values(ascending=True)
```
