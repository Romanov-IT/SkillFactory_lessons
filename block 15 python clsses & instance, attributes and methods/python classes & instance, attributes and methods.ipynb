# python classes & instance, attributes and methods

### object and methods

```python
number = 2.5 #object -> class float
print(number.is_integer()) #method
people = ["Vasiliy", "Stanislav", "Alexandra", "Vasiliy"] #object -> class list
print(people.count("Vasiliy")) #method
```

### classes

```python
number = 2.5  
print(number.__class__) 
people = ["Vasiliy", "Stanislav", "Alexandra", "Vasiliy"]  
print(people.__class__)
```

### instance

```python
class SalesReport():  
    pass
report = SalesReport()
report.__class__
```

```python
class DepartmentReport():

    def add_revenue(self, amount): #method
        if not hasattr(self,'revenues'):
            self.revenues=[] #attribute
        self.revenues.append(amount)
        print(self.revenues)

    def average_revenue(self): #method
            return sum(self.revenues)/len(self.revenues) #float

rep = DepartmentReport()
rep.add_revenue(1500)
rep.add_revenue(4500)
rep.add_revenue(3000)
rep.average_revenue()
```

### init

```python
class DepartmentReport():

    def __init__(self):
         self.revenues = [] #attribute
    
    def add_revenue(self, amount): #method
        self.revenues.append(amount)
        print(self.revenues)
    
    def average_revenue(self): #method
            return sum(self.revenues)/len(self.revenues) #float

rep = DepartmentReport()
rep.add_revenue(1500)
rep.add_revenue(4500)
rep.add_revenue(3000)
rep.average_revenue()
```
