# array functions

```python
import numpy as np
```

```python
arr = np.arange(8)
print(arr)
```

### to change a shape

```python
arr.shape = (2,4)
print(arr)
```

```python
arr_new = arr.reshape((2,4))
print(arr_new)
```

### to change shape of array to order
* order='c' is a default, array is filled in line by line
* order='f' array is filled in by columns

```python
arr_00 = np.arange(10)
arr_00.shape=(2,5)
print(arr_00)
print('-------------')

arr_01 = arr_00.reshape((2,5), order="C")
print(arr_01)
print('-------------')

arr_01 = arr_00.reshape((2,5), order="F")
print(arr_01)
print('-------------')

arr_01 = arr_00.reshape((2,5), order="A")
print(arr_01)
```
