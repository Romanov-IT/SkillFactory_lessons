# array functions

```python
import numpy as np
```

```python
arr = np.arange(8)
print(arr)
```

### to change a shape

```python
arr.shape = (2,4)
print(arr)
```

```python
arr_new = arr.reshape((2,4))
print(arr_new)
```

### to change shape of array to order
* order='c' is a default, array is filled in line by line
* order='f' array is filled in by columns

```python
arr_00 = np.array([[1, 2, 3, 4],
                   [5, 6, 7, 8]
                  ])
print(arr_00)
print('-------------')
arr_01 = arr_00.reshape((4,2), order="F")
print(arr_01)
```

### to transpose array

```python
arr = np.arange(10)
arr.shape = (2,5)
print(arr)

arr_t = arr.transpose()
print(arr_t)
```

### sections of 1d array

```python
arr_02 = np.linspace(0, 1, 9)
print(arr_02)
print('--------------------------')
print(arr_02[2])
print('--------------------------')
print(arr_02[0:2])
print('--------------------------')
print(arr_02[::-1])
```

### sections of 2d array

```python
arr_03 = np.linspace(0, 1, 9).reshape(3,3)
print(arr_03)
print('--------------------------')
print(arr_03[0])
print('--------------------------')
print(arr_03[::-1,-1])
```

### to sort 1d array

* method 1

```python
arr_04 = np.array([12, 23, 56, 1, 4, 3, 17])
print(arr_04)
arr_05 = np.sort(arr_04)
print(arr_05)
```

* method 2

```python
print(arr_04.sort())
print(arr_04)
```

### zero data

```python
arr_06 = np.array([4, -9, 16, -36])
print(arr_06)
roots = np.sqrt(arr_06)
print(roots)
```

### nan is not a number
* np.nan is class 'float', not class 'NoneType'

```python
print(type(None))
print(type(np.nan))
print(None == None)
print(np.nan == np.nan)
print(None is None)
print(np.nan is np.nan)
print(np.nan is None)
```

```python
print(roots)
print(np.isnan(roots))
print(roots[np.isnan(roots)])
roots[np.isnan(roots)] = (9,36)
print(roots)
```
