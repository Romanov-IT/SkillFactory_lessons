```python
covid_df = pd.read_csv('data/covid_df.csv', index_col=False)
```

### .plot()
* x= name of column for axis x
* y= name of column for axis y
* figsize= graph size in inches
* kind= vizualiztion type
* * 'line' by default linear graph
* * 'bar' bar graph
* * 'barh' horizontal bar graph
* * 'hist' histogram
* * 'box' boxplot
* * 'pie' pie graph
* * 'scatter' scatter graph
* grid=False by default grid display
* legend=False be default legend display
* title= graph name
* color

```python
grouped_cases = covid_df.groupby('date')['daily_confirmed'].sum()
grouped_cases.plot(
    kind='line',
    figsize=(17, 8),
    title='Daily incidence graph (line)',
    grid=True,
    lw=1,
    color='red'
);
```

```python
grouped_cases.plot(
    kind='hist',
    figsize=(17, 8),
    title='Daily incidence graph (histogram)',
    grid=True,
    legend=True,
    color='blue',
    bins=10
);
```

```python
grouped_country = covid_df.groupby('country')['confirmed'].last()
grouped_country = grouped_country.nlargest(10)
grouped_country.plot(
    kind='bar',
    grid=True,
    figsize=(17, 8),
    colormap='plasma'
);
```

```python
grouped_country = covid_df.groupby('country')[['confirmed', 'deaths']].last()
grouped_country = grouped_country.nlargest(10, columns='confirmed')
grouped_country.plot(
    kind='bar',
    grid=True,
    figsize=(17, 8)
);
```

```python
grouped_death = covid_df.groupby('country')['deaths'].last()
grouped_death = grouped_death.nlargest(10)
grouped_death.plot(
    kind='bar',
    grid=True,
    ylabel='Deaths number',
    xlabel='Country',
    color='black'
)
```

```python
grouped_vaccines = covid_df.groupby('country')['total_vaccinations'].last()
grouped_vaccines = grouped_vaccines.nsmallest(10)
grouped_vaccines.plot(
    kind='bar',
    grid=True,
);
```
