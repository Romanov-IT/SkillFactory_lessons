### to shuffle numbers

```python
arr = np.arange(6)
print(arr)
arr0 = np.random.shuffle(arr)
print(arr0)
print(arr)
```

```python
arr0 = np.array([1, 2, 3, 4, 5])
arr1 = np.random.permutation(arr0)
print(arr0)
print(arr1)
```

### rabdom choice

```python
arr0 = np.array([1, 2, 3, 4, 5])
arr1 = np.random.choice(arr0, size=(3,3), replace=True)
print(arr1)
arr1 = np.random.choice(arr0, size=(5), replace=False)
print(arr1)
```
