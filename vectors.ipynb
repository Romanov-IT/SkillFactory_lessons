# vectors

```python
import numpy as np
```

### vectors must be same length

### to vectors sum

```python
vec1 = np.array([2, 17, 21, 6])
vec2 = np.array([23, 8, 4, 19])
print(vec1 + vec2)
```

```python
lst1 = [2, 17, 21, 6]
lst2 = [23, 8, 4, 19]
print(lst1 + lst2)

print([x + y for x, y in zip(lst1, lst2)])
```

### vector subtraction

```python
print(vec1 - vec2)
```

### vectors multiplication

```python
vec3 = np.array([2, 4, 6, 8])
print(vec3*5)
print(vec3**2)
print(vec3**0.5)
```

### vectors dividing

```python
vec1 = np.array([2, 17, 21, 6])
vec2 = np.array([23, 8, 4, 19])
print(vec1 / vec2)
print(vec1 // vec2)
print(vec1 % vec2)
print(vec1 / 2)
```

### vector comparison

```python
print(vec1 > vec2)
print(vec3 > 5)
```
