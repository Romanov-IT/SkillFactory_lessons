# vectors

```python
import numpy as np
```

### vectors must be same length

### to vectors sum

```python
vec1 = np.array([2, 17, 21, 6])
vec2 = np.array([23, 8, 4, 19])
print(vec1 + vec2)
```

```python
lst1 = [2, 17, 21, 6]
lst2 = [23, 8, 4, 19]
print(lst1 + lst2)

print([x + y for x, y in zip(lst1, lst2)])
```

### vector subtraction

```python
print(vec1 - vec2)
```

### vectors multiplication

```python
vec3 = np.array([2, 4, 6, 8])
print(vec3*5)
print(vec3**2)
print(vec3**0.5)
```

### vectors dividing

```python
vec1 = np.array([2, 17, 21, 6])
vec2 = np.array([23, 8, 4, 19])
print(vec1 / vec2)
print(vec1 // vec2)
print(vec1 % vec2)
print(vec1 / 2)
```

### vector comparison

```python
print(vec1 > vec2)
print(vec3 > 5)
```

### vectot lenth
* v_len = (x^2 + x^2 + ...)**0.5

```python
vec4 = np.array([3, 4])
v_len = np.sqrt(np.sum(vec4**2))
print(v_len)
```

```python
v_len = np.linalg.norm(vec4)
print(v_len)
```

### distance vectors
* v_dis = ((x - y)^2 + (x - y)^2 + ... )**0.5

```python
vec1 = np.array([0, 3, 5])
vec2 = np.array([12, 4, 7])
v_dis = np.sqrt(np.sum((vec1 - vec2)**2))
print(v_dis)
```

```python
v_dis = np.linalg.norm(vec1 - vec2)
print(v_dis)
```

### scalar product/vector's dot
* v_dot = (x * y) + (x * y) + ...  

```python
vec1 = np.array([0, 3, 5])
vec2 = np.array([12, 4, 7])
v_dot = np.sum(vec1 * vec2)
print(v_dot)
```

```python
v_dot = np.dot(vec1, vec2)
print(v_dot)
```

### vector's statictic functions

```python
vec5 = np.array([1458, 75, 633, 899, 11, 7, 15, 3354])
```

### to get vector's median

```python
mdn = np.median(vec5)
print(mdn)
```

### to get vector's mean

```python
men = np.mean(vec5)
print(men)
```

### to get vector's standart deviation

```python
dev = np.std(vec5)
print(dev)
```

### to get vector's percentile

```python
prc = np.percentile(vec5, 70)
print(prc)
```

### to get vector's min and max

```python
n_mn = np.min(vec5)
n_mx = np.max(vec5)
print('min=', n_mn, 'max=', n_mx)
```
