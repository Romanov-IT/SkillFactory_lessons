# pandas filtering data

```python
import pandas as pd
```

mask is a Series, that includes index that has value of "True" for condition.

```python
melb_data = pd.read_csv("data/melb_data.csv", sep=",")
```

conditions:
* "&" is and
* "|" is or

### filtering with one condition

```python
mask = melb_data["Price"] > 2000000
display(mask)
```

```python
display(melb_data[mask].head(10))
```

```python
display(melb_data[melb_data["Price"] > 2000000].head(10))
```

```python
print(melb_data[melb_data["Rooms"] == 3].shape[0])
display(melb_data[melb_data["Rooms"] == 3].head(5))
```

### filtering with two and more conditions

```python
melb_data[(melb_data["Rooms"] == 3) & (melb_data["Price"] < 300000)].shape[0]
```

```python
melb_data[((melb_data["Rooms"] == 3) | (melb_data["BuildingArea"] > 100)) & (melb_data["Price"] < 300000)].shape[0]
```

### filtering and statistical method

```python
melb_data[melb_data["Type"] == "t"]["Rooms"].max()
```

```python
max_num_rooms = melb_data[melb_data["Type"] == "t"]["Rooms"].max()
melb_data[(melb_data["Type"] == "t") & (melb_data["Rooms"] == max_num_rooms)]
```
