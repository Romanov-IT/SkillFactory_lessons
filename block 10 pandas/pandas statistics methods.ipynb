# pandas statistics methods

```python
import pandas as pd
```

Aggregating method is method, that returns only one value, for each column.

```python
all_agg_methods = pd.DataFrame(
    data = [[".count()", "number of non-empty values"],
            [".mean()", "mean value"],
            [".min()", "minimum value"],
            [".max()", "maximum value"],
            [".var()", "variance value"],
            [".std()", "standard deviation"],
            [".sum()", "sum of values"],
            [".quantile(x)", "quantile of level(x)"],
            [".nunique", "number of unique values"],
            ["non modal method", "non modal method"],
            [".mode()", "modal value"]
           ],
    columns = ["Method", "Description" ]
)
```

```python
display(all_agg_methods)
```
