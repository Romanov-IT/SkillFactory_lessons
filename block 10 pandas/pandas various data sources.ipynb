# pandas various data sources

```python
import pandas as pd
```

### csv is comma-sepatated values

### writing to csv-file
* path_or_buf= path to folder, where data will be written
* sep= separator of data in written file (',' as default)
* decimal= separator of numbers in written file ('.' as default)
* columns= list of columns, to be written to to file (all columns are written as default)
* index= it need to create an additional column with indexes ('True' as default)

```python
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'area': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})

countries_df.to_csv('data/countries.csv', index=False, sep=';')
```

### reading csv-file
* filepath_or_buffer= path to reading file
* sep= separator of data in reading file (',' as default)
* decimal= separator of numbers in written file ('.' as default)
* names= list of column names to be read
* skiprows= number of rows to be skipped

```python
countries_data = pd.read_csv('data/countries.csv', sep=';')
display(countries_data)
```
