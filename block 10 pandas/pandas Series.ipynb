# pandas Series
Series is ordered, changeable, collections of objects, that has associative labels (index).

```python
import pandas as pd
```

### to add series

way with use a list

```python
countries = pd.Series(
    data = ['Англия', 'Канада', 'США', 'Россия', 
            'Украина', 'Беларусь', 'Казахстан'
           ],
    index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ'],
    name = 'countries'
)
display(countries)
```

way with use a dict

```python
countries_00 = pd.Series({
    'UK': 'Англия',
    'CA': 'Канада',
    'US' : 'США',
    'RU': 'Россия',
    'UA': 'Украина',
    'BY': 'Беларусь',
    'KZ': 'Казахстан'},
    name = 'countries'
)
display(countries_00)
```

way without use index

```python
countries_01 = pd.Series(
    ['Англия', 'Канада', 'США', 'Россия', 
     'Украина', 'Беларусь', 'Казахстан'
    ],
    name = 'countries'
)
display(countries_01)
```

### data access 

use .loc

```python
print(countries.loc['RU'])
print('------------------------------')
print(countries.loc[['RU', 'US', "KZ"]])
print('------------------------------')
print(type(countries.loc['RU']))
print('------------------------------')
print(type(countries.loc[['RU', 'US', "KZ"]]))
```

use .iloc

```python
print(countries.iloc[3])
print('------------------------------')
print(countries.iloc[[3, 4, 5]])
print('------------------------------')
print(type(countries.iloc[3]))
print('------------------------------')
print(type(countries.iloc[[3, 4, 5]]))
print('------------------------------')
print(countries.iloc[0:3])
```

# series pracrice

Write the fuction, that takes list 'name' and list 'count' as input. The function returns pd.series.
* use index of item is name
* use value of item is count

```python
names = ['chlorhexidine', 'cyntomycin', 'afobazol'] 
counts = [15, 18, 7]
```

```python
def create_medications(names, counts):
    medications = pd.Series(
        data = counts,
        index = names,
        name = 'medications')
    return medications
```

Write the fuction, that takes pd.series and str 'name' as input. The function returns name and number of items as percent of total number.

```python
medications = create_medications(names, counts)
name = "chlorhexidine"
```

```python
def get_percent(medications, name):
    return (name, (medications[name] / sum(medications[0:]))*100)
get_percent(medications, name)
```
