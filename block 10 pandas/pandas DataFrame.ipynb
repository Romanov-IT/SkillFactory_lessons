```python
import pandas as pd
```

### to add DateFrame

way with use a dict

```python
countries_df = pd.DataFrame({
    'countr':['Англия', 'Канада', 'США', 'Россия', 
              'Украина', 'Беларусь', 'Казахстан'],
    'population':[56.29, 38.05, 322.28, 146.24, 
                  45.5, 9.5, 17.04],
    'area':[133396, 9984670, 9826630, 17125191, 
    603628, 207600, 2724902],
})
display(countries_df)
countries_df.index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ']
display(countries_df)
```

### access to data

```python
display(countries_df.area)
display(countries_df.population)
```

```python
display(countries_df['area'])
display(countries_df['population'])
```

### DataFrame practice

Write the function, that takes three lists as input: "Income", "Expenses" and "Years". The function returns DataFrame.
* use list "Years" as index

```python
income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]
```

```python
def create_companyDF(income, expenses, years):
    result = pd.DataFrame({
        "income":[*income],
        "expenses":[*expenses],
    })
    result.index = [*years]
    return result
```

```python
df = create_companyDF(income, expenses, years)
year = 2025
display(df.index)
```

Write the function, that takes DataFrame table and year sa input. The function returns a difference between income and expenses.
* the fuction must return "None", if year isn't exsist in table

```python
def get_profit(df, year):
    if year in df.index:
        return df.loc[year, 'income'] - df.loc[year, 'expenses']
    return None
display(get_profit(df, 2021))
```
