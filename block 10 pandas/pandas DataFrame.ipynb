# pandas DataFrame
DataFrame is 2 demensial stucture looks like a table, that include rows and columns. Colums is "Series" objects and rows are "Series's" objects. DataFrame also has labels (index), that correspond to each row in the table.

```python
import pandas as pd
```

### to add DateFrame

way with use a dict

```python
countries_df = pd.DataFrame({
    'countr':['Англия', 'Канада', 'США', 'Россия', 
              'Украина', 'Беларусь', 'Казахстан'],
    'population':[56.29, 38.05, 322.28, 146.24, 
                  45.5, 9.5, 17.04],
    'area':[133396, 9984670, 9826630, 17125191, 
    603628, 207600, 2724902],
})
display(countries_df)
countries_df.index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ']
display(countries_df)
```

way with use a list

```python
countries_df_0 = pd.DataFrame(
    data=[['Англия', 56.29, 133396],
          ['Канада', 38.05, 9984670],
          ['США', 322.28, 9826630],
          ['Россия', 146.24, 17125191],
          ['Украина', 45.5, 603628],
          ['Беларусь', 9.5, 207600],
          ['Казахстан', 17.04, 2724902]
         ],
    columns=['country', 'population', 'area'],
    index=['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ'] 
)
display(countries_df_0)
```

### axis

axis=0 axis by rows

```python
countries_df_0.mean(axis=0, numeric_only=True)
```

axis=1 axis by columns

```python
countries_df_0.mean(axis=1, numeric_only=True)
```
