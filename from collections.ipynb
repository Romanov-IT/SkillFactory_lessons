# collections Counter and defaultdict

## Counter

```python
from collections import Counter
```

```python
a = Counter()
a['apple'] += 10
a['orange'] += 7
print(a)
```

### to count items in list

```python
cars = ['red', 'blue', 'black', 'black', 'black', 'red', 'blue', 'red', 'white', 'white', 'white']
cars_count = Counter(cars)
print(cars_count)
print(cars_count['blue'])
```

### to get sum of items

```python
print(sum(cars_count.values()))
```

### to join or subtract lists and etc

```python
cars_moscow = ['black', 'black', 'white', 'black', 'black', 
               'white', 'yellow', 'yellow', 'yellow']
cars_spb = ['red', 'black', 'black', 'white', 'white',
            'yellow', 'yellow', 'red', 'white']
c_moscow = Counter(cars_moscow)
c_spb = Counter(cars_spb)

print(c_moscow)
print(c_spb)
print(c_moscow + c_spb)

c_moscow.subtract(c_spb)

print(c_moscow)
```

### to get list of all/unique items in list and etc

```python
fruits = ['apple', 'apple', 'orange', 'apple', 'banana', 'orange']
c_fruits = Counter(fruits)
print(*c_fruits.elements())
print(list(c_fruits))
```

### to convert to dict or list

```python
print(dict(c_fruits))
print(list(c_fruits))
```

```python
### to get most common items
```

```python
print(c_fruits.most_common())
print(c_fruits.most_common(1))
```

### to clear Counter

```python
print(c_fruits.clear())
```
