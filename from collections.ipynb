# collections Counter and

```python
from collections import Counter
```

```python
a = Counter()
a['apple'] += 10
a['orange'] += 7
print(a)
```

### to count objects in list

```python
cars = ['red', 'blue', 'black', 'black', 'black', 'red', 'blue', 'red', 'white']
cars_count = Counter(cars)
print(cars_count)
```
