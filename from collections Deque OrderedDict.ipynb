# from collections Deque and OrderedDict

```python
from collections import OrderedDict
```

# OrderedDict

```python
data = [('Ivan', 19),('Mark', 25),('Andrey', 23),('Maria', 20)]
name_age = dict(data)
print(name_age)
```

```python
o_name_age = OrderedDict(name_age)
print(o_name_age)
```

### to sort dict

```python
o_name_age = OrderedDict(sorted(data, key=lambda x: x[1], reverse=True))
print(o_name_age)
```

### to add a new item

```python
o_name_age['Danchik'] = 24
print(o_name_age)
```

### to delete item

```python
del o_name_age['Mark']
print(o_name_age)
```
