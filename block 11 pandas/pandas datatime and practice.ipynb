# pandas datetime 

```python
import pandas as pd
```

```python
melb_df = pd.read_csv("data/melb_df.csv", sep = ",")
melb_df.head(5)
```

```python
display(melb_df["Date"])
```

### to chenge order of writing of date

```python
melb_df["Date"] = pd.to_datetime(melb_df["Date"], dayfirst=True)
display(melb_df["Date"])
display(melb_df["Date"][0])
```

### .dt.
* date
* year, month, day 
* hour, minute, second
* dayofweek serial number of week from 0 to 6
* day_name 
* dayofyear serial number of day of year
* quarter 1/4 of year 

```python
years_sold = melb_df["Date"].dt.year
print(years_sold)
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
```

```python
melb_df["MonthSale"] = melb_df["Date"].dt.month
melb_df["MonthSale"].value_counts(normalize=True)
```

### time intervals

```python
delta_days = melb_df["Date"] - pd.to_datetime("2016-01-01")
display(delta_days)
```

```python
display(delta_days.dt.days)
```

```python
melb_df["AgeBuilding"] = melb_df["Date"].dt.year - melb_df["YearBuilt"]
display(melb_df["AgeBuilding"])
display(melb_df["AgeBuilding"].max())
display(melb_df["AgeBuilding"].min())
```

```python
melb_df.drop('YearBuilt', axis=1)
```

```python
melb_df["WeekdaySale"] = melb_df["Date"].dt.dayofweek
display(melb_df["WeekdaySale"])
melb_df.head(3)
```

### pandas datetime practice

```python
ufo = pd.read_csv("data/ufo.csv", sep= ",")
ufo.head(10)
```

```python
ufo["Time"] = pd.to_datetime(ufo["Time"], yearfirst=True)
ufo.head(3)
```

What year had the highest number of UFO sightings in the United States?

```python
ufo["year_sightings"] = ufo["Time"].dt.year
ufo["year_sightings"].value_counts(normalize=True)
```

Find the average time interval (in days) between two consecutive UFO sightings in the state of Nevada (NV).

```python
time_diff_NY = ufo[ufo["State"] == "NY"]["Time"].diff()
time_diff_NY = time_diff_NY.dt.days
print(round(time_diff_NY.mean()))
```
