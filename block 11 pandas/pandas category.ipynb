# pandas category

```python
import pandas as pd
```

# table attribute: 
## categorical
* discrete integer (number of rooms, days, peoples)
* continuous int or float (square of room, price, mass)
## numerical
* name (sex, group, group sex:)
* serial (level, stage, marks)

```python
melb_df = pd.read_csv("data/melb_df_02.csv")
melb_df.head()
```

```python
unique_list = []
for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtype)
    unique_list.append(item)
unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type'] 
).sort_values(by='Num_Unique', ignore_index=True)
display(unique_counts)
```

### to add data type category

```python
display(melb_df.info())
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car', "Suburb"]
max_unique_count = 150
for col in melb_df.columns:
    if melb_df[col].unique().any() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype("category")
```

```python
display(melb_df.info())
```

### attributes of category

```python
print(melb_df['Regionname'].cat.categories)
```

```python
display(melb_df['Regionname'].cat.codes)
```

```python
melb_df["Type"] = melb_df["Type"].cat.rename_categories({
    "u":"unit",
    "t":"towngous",
    "h":"house"
})
display(melb_df["Type"])
```

```python
suburb_index = melb_df["Suburb"].value_counts().nlargest(119).index
melb_df["Suburb"] = melb_df["Suburb"].apply(lambda x: x if x in suburb_index else "other")
melb_df["Suburb"] = melb_df["Suburb"].astype("category")
melb_df.info()
```
