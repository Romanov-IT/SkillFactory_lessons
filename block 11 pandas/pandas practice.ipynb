### delete unnecessary columns

```python
citi.drop(["start station id", "end station id"], axis=1, inplace=True)
```

```python
citi.info()
```

### replace column "birth year" to age of clients, the column name is "age".
* now is "2018"

```python
citi["age"] = citi["birth year"].apply(lambda x: 2018 - x)
citi.drop("birth year", axis=1, inplace=True)
citi.head(5)
```

### how many trips were made by customers over the age of 60?

```python
citi[citi["age"] > 60].shape[0]
```

### create a trip duration attribute
* name of new column is "trip duration"

```python
citi["starttime"] = pd.to_datetime(citi["starttime"])
citi["stoptime"] = pd.to_datetime(citi["stoptime"])
citi.info()
```

```python
citi["trip duration"] = citi["stoptime"] - citi["starttime"]
print(citi["trip duration"].head())
print(citi["trip duration"][3])
```

### create new attribute "weekend"
### how many trips started on weekends?

```python
citi["weekend"] = citi["starttime"].dt.dayofweek.apply(lambda x: 1 if x >= 5 else 0)
citi[citi["weekend"] == 1].shape[0]
```

### create new attribute "time_of_day". the time of day based on the start time of the trip
* between 0 ot 6 - night
* between 7 ot 12 - morning
* between 13 ot 18 - day
* between 19 ot 23 - evening

```python
def day_time(time):
    if 0 <= time < 7:
        return "night"
    elif 7 <= time < 13:
        return "morning"
    elif 13 <= time < 19:
        return "day"
    elif 19 <= time < 24:   
        return "evening"
citi["time_of_day"] = citi["starttime"].dt.hour.apply(day_time)
```

```python
citi["time_of_day"].value_counts()
```
