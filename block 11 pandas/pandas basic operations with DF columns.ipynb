# pandas basic operations with DF columns

```python
import pandas as pd
```

```python
melb_data = pd.read_csv("data/melb_data_ps.csv", sep=",")
print(melb_data.info())
```

### to copy df table

```python
melb_df = melb_data.copy()
display(melb_df.head(5))
```

### to delete columns (.drop() method)
* lables= serial numbers or names of columns, that needs to delete. if they more then one, use a list
* axis=0 delete rows
* axis=1 delete columns
* inplace=True replace original df
* inplace=False returns copy of df. False as default

```python
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head(5)
```

### math operations with columns 

```python
total_rooms = melb_df["Rooms"] + melb_df["Bedroom"] + melb_df["Bathroom"]
display(total_rooms)
```

### to add a new column

```python
melb_df["MeanRoomsArea"] = melb_df["BuildingArea"] / total_rooms
display(melb_df["MeanRoomsArea"])
```

```python
diff_area = melb_df["BuildingArea"] - melb_df["Landsize"]
sum_area = melb_df["BuildingArea"] + melb_df["Landsize"]
melb_df["AreaRatio"] = diff_area / sum_area
display(melb_df["AreaRatio"])
```

```python
melb_df.to_csv("data/melb_df.csv", index=False)
```
