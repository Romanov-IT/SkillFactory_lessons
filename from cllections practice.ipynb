```python
from collections import OrderedDict
from collections import deque
```

```python
temps = [('2000', -4.4), ('2001', -2.5), ('2002', -4.4), ('2003', -9.5)]

def check(lst):
    temp_dict = OrderedDict(sorted(lst, key=lambda x: x[1], reverse=True))
    return temp_dict
print(check(temps))
```
