# Numpy

```python
import numpy as np
```

### Bit

Bit is a position in computer memory, that can keep '0' or '1'.

### Byte

Byte is a group of 8 bits.

### integer

```python
a = np.int8(25)
print(a)
```

### to get info about int

```python
print(np.iinfo(np.int8))
print(np.iinfo(np.int64))
```

### unsigned integer

```python
b = np.uint8(128)
print(b)
```

### to get info about uint

```python
print(np.iinfo(np.uint8))
```

data type don't preserved, if you try to preserve new value of variable

```python
a = np.int8(16)
print(type(a))
a = 32
print(type(a))
```

### to preserve a new value of variable

```python
a = np.int8(16)
print(type(a))
a = np.int8(100)
print(type(a))
```

if you try to adding values in old version of numpy, you get a default data type int64

```python
b = np.int8(12)
print(type(b))
c = 21 + b
print(type(c))
```

if you adding different data type values, but all of them are np.int, 
you get 'senior' data type

```python
d = np.int8(45)
e = np.int16(1)
d += e
print(type(d))
```

### float

```python
f = np.float16(3.14)
print(type(f))
print(f)
```

### to get info about float

```python
print(np.finfo(np.float16()))
```
