```python
from collections import OrderedDict
from collections import deque
from collections import defaultdict
```

```python
temps = [('2000', -4.4), ('2001', -2.5), ('2002', -4.4), ('2003', -9.5)]

def check(lst):
    temp_dict = OrderedDict(sorted(lst, key=lambda x: x[1], reverse=True))
    return temp_dict
print(check(temps))
```

```python
def brackets(line):
    check_lst = deque()
    for i in line:
        if i == '(':
            check_lst.append(i)
        elif i == ')':
            try:
                check_lst.pop()
            except IndexError:
                return False    
    if check_lst:
        return False
    return True
print(brackets("(()())())")) 
```

```python
ratings = [
('Old York', 3.3), ('New Age', 4.6), ('Old Gold', 3.3), ('General Foods', 4.8),
('Belissimo', 4.5), ('CakeAndCoffee', 4.2), ('CakeOClock', 4.2), ('CakeTime', 4.1),
('WokToWork', 4.9), ('WokAndRice', 4.9), ('Old Wine Cellar', 3.3), ('Nice Cakes', 3.9)
]

cafes = OrderedDict(sorted(sorted(ratings, key=lambda x: x[0]), key=lambda x: x[1], reverse=True))
print(cafes)
```

```python
tasks = [(36871, 'office', False),
(40690, 'office', False),
(35364, 'voltage', False),
(41667, 'voltage', True),
(33850, 'office', False)]

def task_manager(tasks):

    urgency_list = defaultdict(deque)
    for server_id, name_server, urgency in tasks:
        if urgency == True:
            urgency_list[name_server].appendleft(server_id)
        else:
            urgency_list[name_server].append(server_id)
    return urgency_list
        
        
task_manager(tasks)
```
